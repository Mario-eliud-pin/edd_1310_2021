```
class Nodo:
    def __init__(self, valor, siguiente =None):
        self.valor=valor
        self.next=siguiente
class circle_list():
    def __init__(self):
        self.primero=None
        self.referencia=None
        self.size=0

    def is_empty(self):
        return self.size==0

    def get_size(self):
        return self.size


    def insert(self, valor):
        if self.is_empty():
            self.primero=self.referencia=Nodo(valor)
            Nodo.next=self.primero
            self.size+=1

        elif valor > self.referencia.valor:
            nuevo = self.referencia
            self.referencia = nuevo.next = Nodo(valor, self.primero)
            self.size += 1

        elif valor > self.referencia.valor:
            curr_node = self.referencia
            self.referencia = curr_node.next = Nodo(valor, self.primero)
            self.size += 1
        elif valor < self.referencia.valor and valor < self.primero.valor:
            curr_node = Nodo(valor)
            curr_node.next = self.primero
            self.primero = curr_node
            self.primero.next = curr_node
            self.size += 1

        elif valor < self.referencia.valor and valor > self.primero.valor:
            nuevo = Nodo(valor)
            curr_node = self.primero
            while nuevo.valor > curr_node.next.valor:
                curr_node = curr_node.next
            nuevo.next = curr_node.next
            curr_node.next = nuevo
            self.size += 1

    def tranversal(self):
        curr_node = self.referencia
        while curr_node.next != self.referencia:
            print(f"{curr_node.next.valor} <-> ", end="")
            curr_node = curr_node.next
        print(self.referencia.valor)
        print("")

    def search(self, valor):
        nuevo = Nodo(valor)
        curr_node = self.referencia
        while nuevo.valor != curr_node.next.valor and curr_node.next != self.referencia:
            curr_node = curr_node.next
        if nuevo.valor == curr_node.next.valor:
            return True
        else:
            return False

    def remove(self, valor):
        nuevo = Nodo(valor)
        curr_node = self.referencia
        if self.search(valor) == False:
            print("No existe el dato que quieres eliminar")
        else:
            while nuevo.valor != curr_node.next.valor and curr_node.next != self.referencia:
                curr_node = curr_node.next
            if curr_node.next.valor == self.referencia.valor:
                curr_node.next = self.referencia.next
                self.referencia = curr_node
                self.size -= 1

            elif curr_node.next.valor == self.primero.valor:
                self.primero = self.primero.next
                self.referencia.next = self.primero
                self.size -= 1

            elif curr_node.next.valor != self.referencia.valor and curr_node.next.valor != self.primero.valor:
                previo = None
                while curr_node.valor != valor and curr_node.next != self.referencia:
                    previo = curr_node
                    curr_node = curr_node.next
                if curr_node.valor == valor:
                    previo.next = curr_node.next
                    self.size -= 1


lc=circle_list()
print(f"esta vacia ? {lc.is_empty()}")
lc.insert(10)
lc.insert(30)
lc.insert(40)
lc.insert(20)
lc.insert(50)
print(f" se encuentra el elemento? {lc.search(20)}")
lc.tranversal()
lc.remove(20)
lc.tranversal()
print(lc.search(60))
print(f"la lista contiene {lc.get_size()} elementos ")
```
