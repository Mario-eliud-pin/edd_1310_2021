```
class NodoDoble:
    def __init__(self, valor, siguiente=None, anterior=None):
        self.data = valor
        self.next = siguiente
        self.prev = anterior

class DoubleLinkedList:
    def __init__(self):
        self.__cabeza = NodoDoble(None, None, None)
        self.__cola = NodoDoble(None, None, None)
        self.__cabeza.next = self.__cola
        self.__cola.prev = self.__cabeza
        self.__size = 0

    def is_empty(self):
        return self.__cabeza.data == None

    def append(self, valor):
        agregar = NodoDoble(valor)
        if self.is_empty():
            self.__cabeza = agregar
            self.__cola = agregar
        else:
            agregar.next = self.__cola
            self.__cola.prev = agregar
            self.__cola= agregar

    def recorrer(self):
        agregar= self.__cabeza
        while agregar != None:
            print(f"{agregar.data} -> ", end=" ")
            agregar = agregar.prev
        print(" ")


    def recorrercola(self):
        agregar= self.__cola
        while agregar != None:
            print(f"{agregar.data} -> ", end=" ")
            agregar = agregar.next

        print(" ")


    def quitarporcola(self, valor):
        pivote = self.__cola
        if self.__cola.data == valor:
            self.__cola = self.__cola.next
            self.quitarporcola(pivote.data)
        else:
            anterior = None
            while pivote.data != valor and pivote != None:
                anterior = pivote
                pivote = pivote.next
                if pivote == valor:
                    anterior.next = pivote.next
                    self.quitarporcola(pivote.data)

    def quitarporcabeza(self, valor):
        pivote = self.__cabeza
        if self.__cabeza.data == valor:
            self.__cabeza = self.__cabeza.prev
            self.quitarporcola(pivote.data)
        else:
            anterior = None
            while pivote.data != valor and pivote.prev != None:
                anterior = pivote
                pivote = pivote.prev
                if pivote.data == valor:
                    anterior.prev = pivote.prev
                    self.quitarporcola(pivote.data)

    def tamaño(self):
        pivote = self.__cabeza
        while pivote != None:
            pivote = pivote.prev
            self.__size = self.__size + 1
        print(f"El tamaño de la lista es de : {self.__size} ")

    def encontrarcabeza(self, valor=None):
        contador = 1
        dato = None
        if valor == None:
            dato = print("no ingresate un valor")
        else:
            pivote = self.__cabeza
            anterior = None
            while pivote.data != valor and pivote.prev != None:
                anterior = pivote
                pivote = pivote.prev
                contador+=1
            if pivote.data == valor:
                dato = dato = print(f"se encuentra en ->: {contador}")
        return dato

    def encontrarporcola(self, valor=None):
        contador = 1
        dato = None

        if valor == None:
            dato = print("no ingresate un valor")
        else:
            pivote = self.__cola
            anterior = None
            while pivote.data != valor and pivote.next != None:
                anterior = pivote
                pivote = pivote.next
                contador+=1
            if pivote.data == valor:
                dato = dato = print(f"se encuentra en->: {contador}")
        return dato


ld = DoubleLinkedList()
ld.is_empty()
print(f"La esta vacia ? { ld.is_empty() } ")
ld.append(10)
ld.append(20)
ld.append(30)
ld.append(40)
ld.append(50)
ld.recorrer()
ld.recorrercola()
ld.quitarporcabeza(30)
ld.recorrer()
ld.quitarporcola(20)
ld.recorrercola()
ld.encontrarcabeza(40)
ld.recorrer()
ld.encontrarporcola(50)
```
