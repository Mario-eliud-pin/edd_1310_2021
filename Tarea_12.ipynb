```
class NodoArbol:
  def __init__(self, value, left=None, right=None ):
    self.data=value
    self.left=left
    self.right=right
arbol=NodoArbol("+",NodoArbol("-",5,4),NodoArbol("*",3,2))
print("recorrido pre :")
print("secuancialmente raiz, izquierda, derecha ")
print(arbol.data)
print(arbol.left.data)
print(arbol.left.left)
print(arbol.left.right)
print(arbol.right.data)
print(arbol.right.left)
print(arbol.right.right)
print("recorrido in")
print("parecido a conciderar a las raices como operadores algebraicos: numero, raiz, numero ")
print(arbol.left.left)
print(arbol.left.data)
print(arbol.left.right)
print(arbol.data)
print(arbol.right.left)
print(arbol.right.data)
print(arbol.right.right)
print("recorrido pos ")
print("izquierda, derecha, raiz")
print(arbol.left.left)
print(arbol.left.right)
print(arbol.left.data)
print(arbol.right.left)
print(arbol.right.right)
print(arbol.right.data)
print(arbol.data)


arbol2=NodoArbol(40,NodoArbol(30,25,35),NodoArbol(50,45,60))
print("recorrido pre")
print(arbol2.data)
print(arbol2.left.data)
print(arbol2.left.left)
print(arbol2.left.right)
print(arbol2.right.data)
print(arbol2.right.left)
print(arbol2.right.right)
print("recorrido in")
print(arbol2.left.left)
print(arbol2.left.data)
print(arbol2.left.right)
print(arbol2.data)
print(arbol2.right.left)
print(arbol2.right.data)
print(arbol2.right.right)
print("recorrido pos")
print(arbol2.left.left)
print(arbol2.left.right)
print(arbol2.left.data)
print(arbol2.right.left)
print(arbol2.right.right)
print(arbol2.right.data)
print(arbol2.data)



```
